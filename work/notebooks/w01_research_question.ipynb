## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import numpy as np

# Load the starter CSV
df = pd.read_csv('/content/ml/data/raw/content_refresh_anonymized.csv')

# 1. Compute total pages and rate of decline
total_pages = len(df)
declining_pages = df[df['trend_direction'] == 'down']
num_declining = len(declining_pages)
pct_declining = (num_declining / total_pages) * 100

# 2. High-impression declining pages (>=1000 impressions in the trailing 90 days)
high_imp_declining = df[(df['trend_direction'] == 'down') & (df['impressions_90d'] >= 1000)]
num_high_imp_declining = len(high_imp_declining)

# 3. Total impressions in dataset and sum of impressions at risk
total_impressions = df['impressions_90d'].sum()
declining_impressions = high_imp_declining['impressions_90d'].sum()
pct_impressions_at_risk = (declining_impressions / total_impressions) * 100

print(f"Total pages in the starter dataset: {total_pages:,}")
print(f"Pages in active decline (trend_direction == 'down'): {num_declining:,} ({pct_declining:.2f}%)")
print(f"High-impression declining pages (>=1,000 impressions in 90d): {num_high_imp_declining:,}")
print(f"Total impressions from declining high-impression pages: {declining_impressions:,} ({pct_impressions_at_risk:.2f}% of total impressions)")
```
